```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid.axislines import SubplotZero
a=0
```

```python
x = np.linspace(-5,5,50000 + 1)
```

```python
f = plt.figure(figsize=(8.5,5))
ax = f.add_subplot(111)
relu = [x if x > 0 else 0 for x in x]
tanh = [np.tanh(x) for x in x]

plt.plot([-5, 5], [0, 0], 'k-', lw=1.0,alpha=0.75,)
plt.plot([-5, 5], [-1, -1], 'k--', lw=1.,alpha=0.68)
plt.plot([-5, 5], [1, 1], 'k--', lw=1.,alpha=0.68)
plt.plot([0, 0], [-2.2, 2], 'k-', lw=1.0,alpha=0.75)

plt.plot(x, 1/(1+ np.exp(-x)), lw=2, label = r'$\sigma$(z)', alpha = 0.97)
tanh = plt.plot(x,tanh, lw=2, label = r'tanh(z)', alpha = 0.97)
ReLU = plt.plot(x,relu, lw=2, label = 'ReLU(z)', alpha = 0.97)

plt.legend(loc='lower right', fontsize='large', framealpha=1, shadow=True, prop={'size': 16})
plt.grid(color='black', alpha=0.20)

plt.ylabel(r'y', fontsize=17)
plt.xlabel(r'z', fontsize=17)
plt.tick_params(direction='in',length=5,width=0.85, labelsize=13.5)
plt.axis([-3.9,3.9, -1.25, 1.3])


plt.savefig('act.png',dpi=30*20)
f
```
